```python
import numpy as np
import matplotlib.pyplot as plt
from SVIT_DATA import SVIT_DATA
```

```python
DATA = SVIT_DATA()
```

```python
keys = DATA.keys()
D = ["CIFAR10", "CIFAR100", "Caltech256"]
A = ["VIT", "SWG"]
H = ["MLP", "SVM"]
```

```python
values = DATA["CIFAR10"]["VIT"]["MLP"]
data_percent = np.arange(1, 6)*20
```

```python
dataset_name = "CIFAR10"
plots, figsize = (1,1), (6, 4)

fig, ax = plt.subplots(*plots, figsize=figsize)
ax.plot(data_percent, DATA[dataset_name]["VIT"]["MLP"], label="ViT - MLP", c="C0")
ax.plot(data_percent, DATA[dataset_name]["VIT"]["SVM"], label="ViT - SVM", c="C1")
ax.plot(data_percent, DATA[dataset_name]["SWG"]["MLP"], label="SWAG - MLP", c="C0", linestyle='dashed')
ax.plot(data_percent, DATA[dataset_name]["SWG"]["SVM"], label="SWAG - SVM", c="C1", linestyle='dashed')
ax.set_title("%s Results"%dataset_name)
ax.set_xticks(data_percent, data_percent)
ax.legend(); ax.set_xlabel("Training Set Percent (%)"); ax.set_ylabel("Test Set Accuracy")
pass
```
